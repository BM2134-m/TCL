```python
# get function- get()
#-If is key is present or inside in dictionry.
#- when you have using get function and key is present or inside in dict output extract particular keys value.

Emp_Data = {'Emp_1': {'Name':'Salahudin','Age': 43,'Salary': 52000},
            'Emp_2': {'Name':'Daksh','Age': 27,'Salary': 28000},
            'Emp_3': {'Name':'Sachin','Age': 47,'Salary': 22900}}
Emp_Data.get('Emp_3')   #- when you have using get function and key is present or inside in dict output extract particular keys value.
```

# Methods to remove the items from dictionary
- Pop function :- pop()
- Popitem fuctions :- popitem ()

```python
# Pop function :- pop()
#- It remove items according to key.
#- What ever item it remove returns output his value only.
#- It is remove particular item user want anything between inside dictionary
#- Only one item remove at time

Emp_Data = {'Emp_1': {'Name':'Salahudin','Age': 43,'Salary': 52000},
            'Emp_2': {'Name':'Daksh','Age': 27,'Salary': 28000},
            'Emp_3': {'Name':'Sachin','Age': 47,'Salary': 22900}}

print(Emp_Data.pop('Emp_2')) #- particular item it remove returns his value only.
print('--------->')
print(Emp_Data)
```

```python
# Popitem function :- pop()
#- popitem () is always remove last item of dictionary.
#- last item of dictionary remove returns the output both key & his value.
#- try to remove items between anything item of dictionary not allowed


Emp_Data = {'Emp_1': {'Name':'Salahudin','Age': 43,'Salary': 52000},
            'Emp_2': {'Name':'Daksh','Age': 27,'Salary': 28000},
            'Emp_3': {'Name':'Sachin','Age': 47,'Salary': 22900}}

print(Emp_Data.popitem('Emp_2')) # try to remove items between anything item of dictionary not allowed error
print('--------->')
print(Emp_Data)
```

```python
# Popitem function :- pop()
#- popitem () is always remove last item of dictionary.
#- last item of dictionary remove returns the output both key & his value.
#- try to remove items between anything item of dictionary not allowed


Emp_Data = {'Emp_1': {'Name':'Salahudin','Age': 43,'Salary': 52000},
            'Emp_2': {'Name':'Daksh','Age': 27,'Salary': 28000},
            'Emp_3': {'Name':'Sachin','Age': 47,'Salary': 22900}}

print(Emp_Data.popitem()) # only remove last item of dictionary and aslo returns both the key& value
print('--------->')
print(Emp_Data)
```
